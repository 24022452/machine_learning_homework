```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import neighbors, datasets
```

```python
#Load data and return a dictionary of iris
iris = datasets.load_iris()
# Dữ liệu đầu vào - features matrix (150,4) - số bông hoa, các chỉ số đo (sepallen, sepalwid, petallen, petalwid)
iris_X = iris.data
# Dữ liệu đầu ra - label tương ứng mảng 1 chiều 150 con số tương ứng hàng của iris_X chỉ gồm 0,1,2 để phân loại hoa
iris_y = iris.target
print (f"Number of classes: {len(np.unique(iris_y))}")
print (f"Number of data point: {len(iris_y)}")

#Some samples of training data
X0 = iris_X[iris_y == 0, :]
print (f"\nSamples of class 0:\n {X0[:5,:]}")

X1 = iris_X[iris_y == 1, :]
print (f"\nSamples of class 1:\n {X1[:5,:]}")

X2 = iris_X[iris_y == 2, :]
print (f"\nSamples of class 2:\n {X2[:5,:]}")
```

## Tách training data(100 point) và test data(50 point) use train_test_split trong sklearn.model_selection

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_X, iris_y, test_size=50)
print (f"Training size: {len(y_train)}")
print (f"Test size: {len(y_test)}")
```

```python
model = neighbors.KNeighborsClassifier(n_neighbors = 1, p = 2)
#p = 2 la chuan khoang cach norm 2
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print ("Print results for 20 test data points")
print (f"Predict label: {y_pred[20:40]}")
print (f"Ground truth: {y_test[20:40]}")
```

## Phương pháp đánh giá (evaluation method)

```python
from sklearn.metrics import accuracy_score
print (f"Accuracy of 1NN: {100*accuracy_score(y_test, y_pred):.2f}%")
```

```python
model = neighbors.KNeighborsClassifier(n_neighbors=10, p = 2)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print (f"Accuracy of 10NN: {100*accuracy_score(y_test, y_pred):.2f}%")
```

```python
model = neighbors.KNeighborsClassifier(n_neighbors=10, p = 2, weights= 'distance')
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print (f"Accuracy of 10NN (1/distance weights): {100*accuracy_score(y_test, y_pred):.2f}%")
```

```python
def myweight(distances):
    sigma2 = .5 # we can change this number
    return np.exp(-distances**2/sigma2)

clf = neighbors.KNeighborsClassifier(n_neighbors = 10, p = 2, weights = myweight)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print ("Accuracy of 10NN (customized weights): %.2f %%" %(100*accuracy_score(y_test, y_pred)))
```

## KNN cho regression

```python
import numpy as np
from mnist import MNIST
import matplotlib.pyplot as plt
from sklearn import neighbors
from sklearn.metrics import accuracy_score
import time

mndata = MNIST('D:/ML/MNIST') # path to your MNIST folder
mndata.load_training()
mndata.load_testing()
X_test = mndata.test_images
X_train = mndata.train_images
y_test = np.asarray(mndata.test_labels)
y_train = np.asarray(mndata.train_labels)
start_time = time.time()
model = neighbors.KNeighborsClassifier(n_neighbors=1, p = 2)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
end_time = time.time()
print (f"Accuracy of 1NN for MNIST: {100*accuracy_score(y_test, y_pred):.2f}%")
print (f"Running time: {(end_time-start_time):.2f}s")
```
